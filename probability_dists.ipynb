```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
```

## Uniform Distribution

```python
uniform_data= stats.uniform.rvs(size=100000,
                               loc=0,
                               scale=10)
uniform_data.max()
sns.kdeplot(uniform_data)
plt.title('Uniform Distribution')
plt.show()
```

`stats.distribution.cdf` is used to determine the probability that an observation drawn from a distribution falls bellow a specified values. in essense, cdf() gives area under the distribution density curve to the left of a certain value on the x axis.

```python
#CDF: 
stats.uniform.cdf(x=2.5, 
                 loc=0,
                 scale=10)
```
